```python
df = pd.read_csv('TS_cmems_reanalysis_temperature_from1987to2019_AdriaticSea.csv', parse_dates=['DATE'])
df.reset_index(inplace=True)

df['Time Series from 1987 to 2019'] = [d.year for d in df.DATE]
df['month'] = [d.strftime('%b') for d in df.DATE]
years = df['Time Series from 1987 to 2019'].unique()

fig, axes = plt.subplots(1, figsize=(20,4), dpi= 100)
sns.boxplot(x='Time Series from 1987 to 2019', y='TEMPERATURE', data=df)

axes.set_title('Temperature at Sea Surface-Trend in Year\n', fontsize=15); 
#plt.savefig('Temperatureat Sea Surface-Trend in Year.png')
plt.show()
plt.close()
```

```python
df['Time Series from 1987 to 2019'] = [d.year for d in df.DATE]
df['Month'] = [d.strftime('%b') for d in df.DATE]
years = df['Time Series from 1987 to 2019'].unique()

fig, axes = plt.subplots(1, figsize=(18,4), dpi= 100)
sns.boxplot(x='Month', y='TEMPERATURE', data=df.loc[~df.month.isin([1987, 2019]), :])

axes.set_title('Temperature at Sea Surface-Trend in Month', fontsize=15); 
#plt.savefig('Temperature at Sea Surface-Trend in Month.png')
plt.show()
plt.close()
```
